```python
import requests
from bs4 import BeautifulSoup
```

```python
page = requests.get("https://en.wikipedia.org/wiki/Main_Page")
print(page.status_code)
```

```python
print(page.content)
```

```python
soup = BeautifulSoup(page.content, "html.parser")
print(soup.prettify())
```
